## Data Cleaning

````
bank_df.loc[bank_df['y'] == 'no', 'y'] = 0
bank_df.loc[bank_df['y'] == 'yes', 'y'] = 1
bank_df.head(12)
````

````
def one_hot_encode(df):
    categorical_columns = df.select_dtypes(include = ['object', 'category']).columns
    df_1e = pd.get_dummies(df[categorical_columns])
    df = df.drop(categorical_columns, axis = 1)
    df = df.join(df_1e)
    return df

bank_df_1e = one_hot_encode(bank_df)
bank_df_1e.head(12)
````

````
bank_df_1e.shape
````

````
X = bank_df_1e.drop(['y_0','y_1'], axis = 1)
X = X.to_numpy()
X
````

````
y = bank_df_1e[['y_0', 'y_1']]
y = y.to_numpy()
y
````

````
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 12)
````

## Data Modeling

````
K.clear_session()
model = Sequential()

model.add(Dense(5, activation = 'relu', input_shape = (51, )))
model.add(Dense(5, activation = 'relu'))
model.add(Dense(2, activation = 'softmax'))

model.compile(loss = tf.keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error"),
              optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001,
                                                      rho=0.90,
                                                      epsilon=1e-07,
                                                      decay=0),
              metrics = [tf.keras.metrics.MeanSquaredError(name="mean_squared_error", dtype=None)])
````

````
experiment = model.fit(X_train, y_train, epochs = 8,
              batch_size = 4, validation_split = 0.30)
````

````
experiment.history.keys()
%matplotlib inline
plt.style.use('fivethirtyeight')
pd.DataFrame(experiment.history).plot(figsize=(7, 5), ls='-')
plt.grid(True)
plt.gca().set_ylim(0, 2)
plt.show();
````
